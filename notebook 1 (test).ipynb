```python
words = open("angry.txt").read().split()
```

```python
initials = {}

# look at the first letter of each word
# check to see if there's already an entry in initials
# if the first letter is not already a key: add it with an empty list
# if it is, then add the word to the (list) value

for item in words:
    first_let = item[0].lower()
    if first_let not in initials:
        initials[first_let] = []
    initials[first_let].append(item)
    #print(item, first_let, initials)
    #print()
```
